## Class and Objects

```python
class A:

    def add():

        a = 10
        b = 20
        print(a+b)

    def subtract():

        a = 100
        b = 20
        print(a-b)

obj = A

obj.add()
obj.subtract()
```

```python
class B:

    def vowel():

        letter = input("Enter Any Letter")

        if(letter in "aeiou"):
            print("Vowel")
        else:
            print("Consonent")

    def number():

        num = int(input("Enter the Number: "))

        if(num>0):
            print("Positive")
        elif(num<0):
            print("Negative")
        elif(num==0):
            print("Zero")
        else:
            print("Invalid Input... Enter a Number")


obj = B
obj.vowel()
obj.number()
    
```

```python
class C:

    def fact():

        print("Program for Factorial")

        num = int(input("Enter a Number:"))

        fact = 1

        for i in range(1,num+1):
            fact *= i

        print(fact)

    def perfect():

        print("Program for Perfect Number...")

        num = int(input("Enter Any Number:"))

        total = 0

        for i in range(1,num):

            if(num%i==0):

                total+=i
        if(total == num):
            print(f"{num} is an Perfect Number.")
        else:
            print(f"{num} is not an Perfect Number.")

obj = C

obj.fact()
obj.perfect()
```
