# Polymorphism

# Over-ridding

```python
# No return No Argument
class A:
    def task(self):

        a = 10
        b = 20
        c = a+b
        print(c)

class B(A):

    def task(self):
        a = 100
        b = 20
        c = a-b
        print(c)

obj = B()

obj.task()
    
        
```

```python
# Only Argument
class A:
    def task(self,a,b):

        c = a+b
        print(c)

class B(A):

    def task(self,a,b):

        c = a-b
        print(c)

obj = B()

obj.task(10,8)
    
        
```

```python
# Only Return
class A:
    def task(self):

        a = 10
        b = 20
        c = a+b
        return c

class B(A):

    def task(self):
        a = 100
        b = 20
        c = a-b
        return c

obj = B()

obj.task()
    
        
```

```python
# Return Plus Argument
class A:
    def task(self,a,b):
        self.a = a
        self.b = b
        c = self.a + self.b
        return c

class B(A):

    def task(self,a,b):
        c = a-b
        return c

obj = B()

obj.task(100,80)
    
        
```
