# Python Questions

# 1. 10 Questions on List

```python
# Create a list of squares from 1 to 10.
a = [i*i for i in range (1,101)]
a
```

```python
# From a list of numbers, make a new list containing only even numbers.
a = [1,2,3,4,5,6,7,8,9,13,14,16,18,19]
b = [i for i in a if i%2==0 ]
b
```

```python
# Get the first letter of each word.
a = ['Cat','Dog','Lion']
b = [i[0] for i in a]
b
```

```python
# Create a list of numbers divisible by 3 from 1 to 50.
b = [i for i in range(1,52) if i%3==0]
b
```

```python
# Convert to Upper CASE
fruits = ["apple", "banana", "mango"]
b = [i.upper() for i in fruits]
b
```

```python
# Replace negative numbers with 0
nums = [3, -1, 5, -7, 9]
b = [x if x>=0 else 0 for x in nums]
b
```

```python
# Extract vowels from a string
str = 'hello world'
vowel = [i for i in str if i in 'aeiou']
vowel
```

```python
# Names starting with 'A'
names = ["Amaan", "Rohit", "Arman", "Bilal"]
b = [i for i in names if i.startswith('A')]
b
```

```python
# Odd numbers from 1 to 20
b = [i for i in range(1,21) if i%2==1]
b
```

# 2. 10 Questions in Dictionary

```python
# Dictionary of numbers and their squares (1–10)
b = {i:i*i for i in range(1,11)}
b
```

```python
# Convert two lists into a dictionary
keys = ["name", "age", "city"]
values = ["Amaan", 21, "Mumbai"]
merge = {k:v for k,v in zip(keys,values)}
merge
```

```python
# Character Count
text = "hello"
count = {ch:text.count(ch) for ch in text}
count
```

```python
# Keep only even values
data = {1: 3, 2: 4, 3: 10, 4: 7}
even = {k:v for k,v in data.items() if k %2 ==0}
even
```

```python
# Swap keys and values
d = {"a": 1, "b": 2, "c": 3}
new = {v:k for k,v in d.items()}
new
```

```python
# Words and their lengths
words = ["apple", "banana", "kiwi"]
d = {word:len(word) for word in words}
d
```

```python
# Multiples of 5 (1–20) mapped to their squares
multiples = {i:i*i for i in range(1,21) if i%5==0}
multiples
```

```python
# Map numbers to "even" or "odd"
d = {x:('even' if x%2==0 else 'odd') for x in range(1,101)}
d
```

```python
# Add +5 bonus marks to each student
marks = {"Aman": 80, "Rohit": 75, "Sara": 90}
bonus = {name:score+5 for name,score in marks.items()}
bonus
```

```python
# Convert list of tuples to dictionary
pairs = [("a", 1), ("b", 2), ("c", 3)]
dict_from_tuples = {k: v for k, v in pairs}
dict_from_tuples
```

# 3. 5 Questions on Lambda



```python
# a lambda function to add 10 to a given number.
add_10 = lambda x: x+10
add_10(20)
```

```python
# Lambda to return square of a number
square = lambda x: x*x
square(5)
```

```python
# Double all numbers using lambda
nums = [1,2,3,4,5]
double = list(map(lambda x:x*2,nums))
double
```

```python
# Filter even numbers using lambda
evens = list(filter(lambda x:x%2==0,nums))
evens
```

```python
# Sort list of tuples by second value using lambda
pairs = [(1, 3), (4, 1), (2, 2)]
sort = sorted(pairs,key = lambda x: x[1])
sort
```

# 4. 5 Questions on Filter

```python
# Extract even numbers
nums = [1, 2, 3, 4, 5, 6]
evens = list(filter(lambda x:x%2==0,nums))
evens
```

```python
# Numbers greater than 10
nums = [5, 11, 3, 15, 20, 7]
greater = list(filter(lambda x: x>10,nums))
greater
```

```python
# Names starting with "A"
names = ["Amaan", "Rohit", "Arman", "Bilal", "Aisha"]
b = list(filter(lambda x:x[0] == 'A',names))
b
```

```python
# Positive numbers only
nums = [-3, 7, 0, -1, 9, 12]
b = list(filter(lambda x:x>0,nums))
b
```

```python
# Words longer than 4 characters
words = ["apple", "cat", "banana", "dog", "kiwi"]
b = list(filter(lambda x: len(x)>=4,words))
b
```

# 5. 5 Questions on Map

```python
# Square every number
nums = [1, 2, 3, 4]
squares = list(map(lambda x:x*x,nums))
squares
```

```python
# Convert numbers to strings
del str
nums = [10, 20, 30]
string = list(map(lambda x:str(x),nums))
string
```

```python
# Convert all words to uppercase
words = ["apple", "banana", "kiwi"]
upper = list(map(lambda word: word.upper(),words))
upper
```

```python
# Get lengths of each word
words = ["hi", "hello", "python"]
lengths = list(map(lambda w: len(w), words))
lengths
```

```python
# Add elements of two lists
a = [1, 2, 3]
b = [4, 5, 6]
summed = list(map(lambda x,y:x+y,a,b))
summed
```
