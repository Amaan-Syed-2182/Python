# ATM Machine

```python
balance = 1000
Pin = 1234

pin = int(input("Enter your Pin: "))

if pin == Pin:
    print('''Enter following Number as per Choice:
            1 : check balance
            2 : Deposit
            3 : Withdrawal ''')
    choice = int(input("Enter Choice Number: "))

    if(choice == 1):
        print(f"Your Current Balance is: {balance}")
    elif(choice == 2):
        deposit = int(input("Enter Deposit Value: "))
        balance += deposit
        print(f"Your Updated Balance is: {balance}")
    elif(choice == 3):

        withdraw = int(input("Enter Withdrwal Amount: "))
        if(withdraw > balance):
            print(f"Sorry Insufficient Balance....")
            
        elif(balance >= withdraw):

            if(withdraw <= 0):

                print(f"Invalid Withdrawal Amount Entered")
                
            else: 
                balance -= withdraw
    
                print(f'''Please Collect your {withdraw}rs.
                Transaction Succesfull!!!''')
                print(f"Remaining Balance: {balance}")

        else:
            print(f"Invalid Withdrwal Amount Entered")
        
```
