# Constructor and Self

```python
# Code with Constructor
class A:
    def __init__(self):
        a = 10
        b = 20
        c = a+b
        print(c)
obj = A()
```

# Self argument with Inheritence

```python
class A:
    def input(self):
        self.a = 10
        self.b = 20

class B(A): 
    def sum(self):
        self.c = self.a + self.b

        print(self.c)
obj = B()
obj.input()
obj.sum()
```

# Practice for Inheritence

```python
# 1. mark sheets using Inheritence

class A:

    def input(self):

        self.maths = int(input("Enter Maths Marks: "))
        self.science = int(input("Enter Science MArks: "))
        self.social = int(input("Enter Social Science Marks: "))
        self.hindi = int(input("Enter Hindi Marks: "))
        self.english = int(input("Enter English Marks: "))

        if(-1<self.maths<101 and -1<self.science<101 and -1<self.social<101 and -1<self.hindi<101 and -1<self.english<101):
            self.sum = self.maths+self.science+self.social+self.hindi+self.english
        else:
            print(f"Invalid Marks Entered... Value Between 0 and 100 are Valid.")

class B(A):

    def total(self):
        print(f"Maths marks out of 100: {self.maths}")
        print(f"Science marks out of 100: {self.science}")
        print(f"Social marks out of 100: {self.social}")
        print(f"Hindi marks out of 100: {self.hindi}")
        print(f"English marks out of 100: {self.english}")

        print(f"Total marks out of 500: {self.sum}")

class C(B):

    def percentage(self):

        self.per = (self.sum/500)*100

        print(f"Total Percentage are: {self.per:.2f}%")

obj2 = C()

obj2.input()
obj2.total()
obj2.percentage()
```

```python
# 2. count digit and sum of digits of an number using Inheritence

class A:

    def input(self):

        self.num = int(input("Enter Any Number: "))

class B(A):

    def count(self):

        digit_count = 0

        original = self.num

        while(original>0):
            
            digit_count += 1

            original //= 10
        
        print(f"Number of Digits in {self.num}: {digit_count}")

class C(A):

    def sum(self):

        total = 0

        original = self.num

        while(original > 0):
            
            digit = original%10

            total += digit

            original //= 10

        print(f"Sum of digits of {self.num} is: {total}")

obj1 = B()
obj1.input()
obj1.count()

obj2 = C()
obj2.input()
obj2.sum()

            

        
```
