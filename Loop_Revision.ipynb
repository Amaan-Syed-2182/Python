```python
# Write a program to find LCM of two numbers.

n1 = int(input("Enter first Number: "))
n2 = int(input("Enter second number: "))
greater = 0
lcm = 0

if(n1>n2):
    greater = n1
else:
    greater = n2

for i in range(greater,n1*n2+1):
    if(i%n1==0 and i%n2==0):
        lcm+=i
        break
print(lcm)
```

```python
# Write a program to check whether a number is an Armstrong number or not.

n = int(input("Enter any number: "))

length = len(str(n))

original = n

total = 0

for i in range(1,length+1):
    digit = n%10
    total += digit ** length
    n//=10

if(original == total):
    print("Armstrong")
else:
    print("Not Armstrong")
```

```python
# Write a program to check whether a number is Strong number or not

n = int(input("Enter any Number: "))
original = n
length = len(str(n))
total = 0

for i in range(1,length+1):
    digit = n%10
    fact = 1
    for j in range(1,digit+1):
        fact *= j
    total += fact
    n//=10
if(total == original):
    print("Strong")
else:
    print("Not Strong")
```
