```python
# 2. Write a program that takes three numbers as input and prints the largest of the three.

n1 = int(input("Enter First Number: "))
n2 = int(input("Enter Second Number: "))
n3 = int(input("Enter Third Number: "))

if(n1>n2 and n1>n3):
    print(f"{n1} is Largest.")
elif(n2>n1 and n2>n3):
    print(f"{n2} is Largest.")
elif(n3>n1 and n3>n2):
    print(f"{n3} is Largest.")
else:
    print(f"All Three are equals.")
    
```

```python
# 3. Write a program to check if a given year is a leap year. A leap year is divisible by 4 but not by 100
#    unless it is also divisible by 400.

year = int(input("Enter the Year: "))

if(year>=2000):
    if(year%4 == 0):
        print(f"{year} is a Leap Year.")
    else:
        print(f"{year} is Not a Leap Year.")
elif(year<2000):
    if(year%400 == 0):
        print(f"{year} is a Leap Year.")
    else: 
        print(f"{year} is Not a Leap Year.")
else:
    print("Invalid Year Entered.")

        
```

```python
'''4 Write a program that takes a percentage (integer) as input and prints the corresponding grade based
on the following criteria:
>= 90: Grade A
>= 80: Grade B
>= 70: Grade C
>= 60: Grade D
< 60: Grade F '''

percent = int(input("Enter your Percentage: "))

if(90<=percent<=100):
    print("Grade A")
elif(80<=percent<=89):
    print("Grade B")
elif(70<=percent<=79):
    print("Grade C")
elif(60<=percent<=69):
    print("Grade D")
elif(0<=percent<=59):
    print("Grade F")
else:
    print("Invalid Percentage")

```

```python
# 5.Write a program that checks if a given letter is a vowel (a, e, i, o, u) or a consonant.

letter = input("Enter a Letter: ")

if(len(letter) == 1):
    if(letter.lower() in 'aeiou'):
        print("Vowel")
    else:
        print("Consonant")
else:
    print("Enter a Letter only.")
```

```python
# 6.Write a basic calculator program that takes two numbers and an operator (+, -, *, /) as input and
#    performs the specified operation. Print the result based on the operation.

print("Welcome to Calculator...")
print('''Enter Following for Specified Operations 
        + Addition
        - Subtraction
        * Multiplication
        / Division''')

a = int(input("Enter First Number: "))
b = int(input("Enter Second Number: "))
operator = input("Enter The Operation: ")

if(operator == '+'):
    print(a+b)
elif(operator == '-'):
    print(a-b)
elif(operator == '*'):
    print(a*b)
elif(operator == '/'):
    print(a/b)
else:
    print("Enter a Valid Operation")
```

```python
# 9 Write a program that takes three sides of a triangle as input and checks if those sides form a valid
# triangle. A triangle is valid if the sum of any two sides is greater than the third side.
# Check conditions like a + b > c, b + c > a, and a + c > b.

side1 = int(input("Enter side 1: "))
side2 = int(input("Enter side 2: "))
side3 = int(input("Enter Side 3: "))

if(side1 + side2 > side3 and side2 + side3 > side1 and side3 + side1 > side2):
    print("Valid Triangle")
else:
    print("Invalid Triangle")
```

```python
# Q10 Write a program that calculates the Body Mass Index (BMI) based on user input for weight (in
# kilograms) and height (in meters). Then categorize the BMI into:
# Underweight (BMI < 18.5)
# Normal weight (18.5 <= BMI < 24.9)
# Overweight (25 <= BMI < 29.9)
# Obesity (BMI >= 30)
# Use the formula: BMI = weight / (height ** 2)

weight = int(input("Enter Weight(kg): "))
height = float(input("Enter Height(Feet): "))
height = height * 0.3048

bmi = weight / (height ** 2)

if(bmi<18.5):
    print("Underweight")
elif(18.5 <= bmi <= 24.9):
    print("Normal Weight")
elif(25 <= bmi <= 29.9):
    print("Overweight")
else:
    print("Obesity")
```
