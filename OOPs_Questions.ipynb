```python
# Write a program to create a class Rectangle to find area and perimeter.

class rectangle:

    def __init__(self,length,width):
        self.length = length
        self.width = width

    def area(self):
        print(f"Area for Rectangle is: {self.length*self.width}.")

    def peri(self):
        print(f"Perimeter is: {2*(self.length+self.width)}.")

obj = rectangle(10,10)
obj.area()
obj.peri()
```

```python
# Create a class BankAccount with methods deposit, withdraw, and display.

class bankaccount:

    def __init__(self,name,acc_num,balance):
        self.name = name
        self.acc_num = acc_num
        self.balance = balance

    def deposit(self):
        deposit = int(input("Enter Deposit Amount:"))

        self.balance += deposit

        print(f"Money Deposited Succesfully!!!")
        print(f"Updated Balance: {self.balance}")

    def display(self):

        print(f"Account Number: {self.acc_num} is Owned by Mr.{self.name} with Balance {self.balance}RS.")

    def withdraw(self):

        amount = int(input("Enter Withdrawal Amount: "))

        if(amount>0):
            
            if(amount>self.balance):

                print(f"Sorry!!! Insuffient Funds...")
            
            else:

                self.balance -= amount

                print(f"Please Collect your Cash!!!")
                print(f"Transaction Succesfull!!!")
                print(f"Remeaning Funds are {self.balance}RS.")

        else:

            print(f"Withdrawal Amount cannot be Zero or Negative.")


obj = bankaccount("Amaan",101,1000)

obj.display()
obj.deposit()
obj.withdraw()          
```

```python
# Create a Class Employee to calculate Salary using Inheritence

class employee:

    def __init__(self, name, emp_id):
        self.name = name
        self.emp_id = emp_id

    def details(self):
        print(f"ID: {self.emp_id} | Name: {self.name}.")


class salary(employee):

    def __init__(self, name, emp_id, salary_amount):
        super().__init__(name, emp_id)
        self.salary_amount = salary_amount

    def calculator(self):
        
        basic = self.salary_amount * 0.50
        hra = self.salary_amount * 0.40
        pf = self.salary_amount * 0.12
        
        
        gross_salary = self.salary_amount - pf

        print(f"Basic Salary: {basic:.0f}rs")
        print(f"HRA: {hra:.0f}rs")
        print(f"PF: {pf:.0f}rs")
        print(f"Gross Salary: {gross_salary:.0f}rs")



obj = salary("Amaan", 101, 100000)
obj.details()
obj.calculator()
```

```python
# Create a class shape and derive classes like circle and rectangle

class Shape:

    def __init__(self, color):
        self.color = color

    def display_color(self):
        print(f"Shape Color: {self.color}")


class Circle(Shape):

    def __init__(self, color, radius):
        super().__init__(color)
        self.radius = radius

    def calculate_area(self):
        pi = 3.14159
        area = pi * self.radius * self.radius
        print(f"Circle Area: {area:.2f}")


class Rectangle(Shape):

    def __init__(self, color, width, height):
        super().__init__(color)
        self.width = width
        self.height = height

    def calculate_area(self):
        area = self.width * self.height
        print(f"Rectangle Area: {area:.2f}")


my_circle = Circle("Blue", 5)
my_circle.display_color()
my_circle.calculate_area()

print("-" * 25)

my_rectangle = Rectangle("Green", 4, 6)
my_rectangle.display_color()
my_rectangle.calculate_area()
```

```python
# Create a class Library to store book details and display them.

class Book:

    def __init__(self, title, author, isbn):
        self.title = title
        self.author = author
        self.isbn = isbn

    def display_info(self):
        print(f"Title: {self.title} | Author: {self.author} | ISBN: {self.isbn}")


class Library:

    def __init__(self):
        self.books = []

    def add_book(self, book):
        self.books.append(book)

    def display_all_books(self):
        if not self.books:
            print("The library is empty.")
            return
        
        print("--- Library Inventory ---")
        for book in self.books:
            book.display_info()


book1 = Book("The Hobbit", "J.R.R. Tolkien", "9780007487288")
book2 = Book("1984", "George Orwell", "9780451524935")

my_library = Library()

my_library.add_book(book1)
my_library.add_book(book2)

my_library.display_all_books()
```
